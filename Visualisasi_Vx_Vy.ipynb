```python
import numpy
import pandas as pd
```

```python
#feature = pd.read_excel('/content/input_big.xlsx')
label = pd.read_excel('/content/data_big_4.xlsx', sheet_name='Sheet2')
data_asli = pd.read_excel('/content/data_big.xlsx', sheet_name='Sheet2')
```

```python
data_asli.head()
```

```python
data_complete = []
f = 0
for i in range(192):
  frame= data_asli.iloc[f:f+154]
  frame = frame.drop(['koor_z'],axis=1)
  frame = frame.drop(['Vz','Temp','source','Jendela','AC',
                      'sudut angin','Nilai Kecepatan Angin','Dry Bulb Temperature',
                      'Fan 1','Fan 2','Fan 3','ACH_jendela','ACH_fan','ACH_total'],axis=1)
  data_complete.append(frame)
  f=+154
```

```python
data_sampled= []
f = 0
for i in range(192):
  df = label.iloc[f:f+30]
  df = df.drop(['Temp'],axis=1)
  data_sampled.append(df)
  f=+30
```

```python
data_sampled[0]
```

```python
import plotly.graph_objects as go
from plotly.figure_factory import create_quiver
def show_image(Y,i):
  x = Y['koor_x']
  y = Y['koor_y']
  u = Y['Vx']
  v = Y['Vy']
  fig = create_quiver(x,y,u,v, scale=500, arrow_scale=i)
  return fig.show()
```

```python
show_image(data_complete[10],0.1)
```

```python
show_image(data_sampled[100],0.1)
```
